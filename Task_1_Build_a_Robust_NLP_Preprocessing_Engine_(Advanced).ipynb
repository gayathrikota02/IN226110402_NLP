# **Task 1: Conceptual Understanding**

**1. What is the difference between "Love" and "love" in NLP?**


In Natural Language Processing, the difference between “Love” and “love” that they are case sensitivity. If text is not normalized, the system treats them as two distinct tokens. For example, “Love” with a capital letter might appear at the beginning of a sentence or as a proper noun, while “love” in lowercase is usually the common word. Without normalization, the model may fail to connect the two, which can affect tasks such as sentiment analysis, word frequency counts, or even search queries. This is why preprocessing steps like converting all text to lowercase are often applied to ensure consistency and avoid misinterpretation.

**2. What happens if stopwords are not removed?**





Stopwords are common words such as 'the', 'is', 'at', 'on', and 'and' that occur frequently in language but usually carry little meaning on their own. If stopwords are not removed, they can fill a dataset with unnecessary, irrelevant, or overlapping information and force models to process unnecessary information. This increases the dimensionality of the text data, slows down training, and reduces efficiency. In tasks like keyword extraction or topic modeling, the presence of stopwords can dilute meaningful patterns, making it harder to identify the core themes of the text. Essentially, leaving stopwords in the dataset can make NLP models less focused and less effective.

**3. Provide two real-world scenarios where removing stopwords can be harmful.**


Although stopword removal is a common preprocessing step, there are situations where it can be harmful. One example is in legal or contractual documents, where words such as 'shall', 'must', and 'not' are critical. Removing them could completely alter the intent of the text—for instance, “You must not share confidential data” would lose its force if 'must' and 'not' were removed. Another example is in search queries for information retrieval systems. A query like “to be or not to be” relies heavily on stopwords to preserve its meaning.

**4. Explain the difference between stemming and lemmatization.**

Stemming and lemmatization are both techniques used to reduce words to their base form, but they differ in approach and accuracy. Stemming is a crude method that simply chops off word endings to produce a root form, which may not always be a valid word. For example, “studies” might become “studi.” Lemmatization, on the other hand, uses vocabulary and grammar rules to return the meaningful base form, ensuring that the result is always a valid word. For instance, “studies” would correctly become “study.” Stemming is faster and useful when exact meaning is not critical, while lemmatization is more accurate and preferred in tasks that require semantic precision, such as machine translation or sentiment analysis.

# **Task 2:Build Advanced Preprocessing Function**

```python
def reduce_repeats(word: str) -> str:
    result = []
    for char in word:
        # If the last two characters are the same as current, skip
        if len(result) >= 2 and result[-1] == char and result[-2] == char:
            continue
        result.append(char)
    return "".join(result)


def preprocess_text(text: str) -> str:
    # 1. skip URLs and email-like patterns
    words = text.split()
    cleaned_words = []
    for word in words:
        if word.startswith("http") or word.startswith("www") or "@" in word:
            continue
        cleaned_words.append(word)
    text = " ".join(cleaned_words)

    # skip numbers
    text = "".join([ch for ch in text if not ch.isdigit()])

    # lowercase
    text = text.lower()

    # Handle repeated characters
    words = text.split()
    words = [reduce_repeats(word) for word in words]

    # Skip short tokens (≤ 2), except 'no' and 'not'
    meaningful = {"no", "not"}
    words = [word for word in words if len(word) > 2 or word in meaningful]

    # Remove extra spaces
    text = " ".join(words).strip()
    return text

sample_message = "WOW!!! This is soooo goooood!!! Visit http://example.com now. I have 3 dogs."
print(preprocess_text(sample_message))
```


# **Task 3: Stress Testing**

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i, sentence in enumerate(sample_inputs, start=1):
    cleaned_sentence = preprocess_text(sentence)
    cleaned_tokens = cleaned_sentence.split()
    print(f"Test {i}")
    print("Original Text :", sentence)
    print("Cleaned Tokens:", cleaned_tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```

# **Task 4: Token Analytics**

1. Which sentence had the most noise?
2. Which sentence retained the most meaningful tokens after cleaning?

```python
analytics = []

for i, sentence in enumerate(sample_inputs, start=1):
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    analytics.append({
        "index": i,
        "original": sentence,
        "cleaned": cleaned_sentence,
        "tokens": tokens,
        "total": total_tokens,
        "unique": unique_tokens,
        "avg_len": round(avg_token_length, 2)
    })

for item in analytics:
    print(f"Test {item['index']}")
    print("Original Text :", item['original'])
    print("Cleaned Tokens:", item['tokens'])
    print("Cleaned Sentence:", item['cleaned'])
    print("Total Tokens   :", item['total'])
    print("Unique Tokens  :", item['unique'])
    print("Average Length :", item['avg_len'])
    print("-" * 50)

# Which sentence had the most noise? (fewest tokens after cleaning)
most_noise = min(analytics, key=lambda x: len(x['tokens']))
print("Sentence with most noise removed:", most_noise['original'])

# Which sentence retained the most meaningful tokens? (most tokens after cleaning)
most_meaningful = max(analytics, key=lambda x: x['total'])
print("Sentence with most meaningful tokens:", most_meaningful['original'])
```

# **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []
for item in analytics:
    all_tokens.extend(item['tokens'])

# Count frequency of each token
token_counts = Counter(all_tokens)

# Top 10 most frequent words
top_10 = token_counts.most_common(10)

# sort by frequency ascending and take first 5
least_5 = sorted(token_counts.items(), key=lambda x: x[1])[:5]

print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word}: {freq}")

print("\nTop 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word}: {freq}")
```


# Task 6: Build Full Pipeline **bold text**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        # preprocess
        cleaned = preprocess_text(text)
        clean_sentences.append(cleaned)

        tokens = cleaned.split()
        all_tokens.extend(tokens)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

pipeline_output = full_pipeline(sample_inputs)

print("All Tokens:", pipeline_output["tokens"])
print("\nCleaned Sentences:")
for sent in pipeline_output["clean_sentences"]:
    print("-", sent)
```
