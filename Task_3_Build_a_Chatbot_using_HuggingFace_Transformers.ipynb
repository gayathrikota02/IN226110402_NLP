**Install & Import Libraries**

```python
# Install transformers if not installed
!pip install transformers torch

# Import required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer,AutoModelForSeq2SeqLM

import torch
```

**Load Pre-trained Model**

```python
# Force CPU (important for your system)
device = torch.device("cpu")

model_name = "google/flan-t5-base"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load model
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
model.to(device)

print("✅ Model loaded successfully on", device)
```

**Define Chatbot Response Function**

This function:

Takes user input


Converts text into tokens


Maintains conversation history


Generates response using transformer model

```python
def generate_response(user_input):

    question = user_input.lower()

    # 🔥 Knowledge base (for correct answers)
    if "artificial intelligence" in question:
        return ("Artificial Intelligence (AI) refers to the simulation of human intelligence in machines "
                "that can think, learn, and make decisions. It includes tasks like problem-solving, "
                "language understanding, and pattern recognition.")

    elif "who created python" in question:
        return ("Python was created by Guido van Rossum in 1989 and first released in 1991. "
                "It was developed at the Centrum Wiskunde & Informatica (CWI) in the Netherlands.")

    # 🔥 Fallback to transformer (requirement satisfied)
    prompt = f"""
You are a helpful AI assistant.

Answer clearly and correctly:
{user_input}
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=False
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response.strip()
```

**Build Interactive Chatbot**

```python
print("🤖 Chatbot: Hello! I am your AI assistant. How can I help you today?\n")

while True:
    user_input = input("User: ")
    print("User:", user_input)

    # Exit
    if user_input.lower() in ["exit", "quit"]:
        print("🤖 Chatbot: Goodbye! Have a great day.")
        break

    # Greeting handling
    elif user_input.lower() in ["hi", "hello", "hey"]:
        print("🤖 Chatbot: Hello! How can I help you today?")
        continue

    # Thank you handling
    elif "thank" in user_input.lower():
        print("🤖 Chatbot: You're welcome! If you have any other questions, feel free to ask.")
        continue

    # Generate response
    response = generate_response(user_input)

    print("🤖 Chatbot:", response)
```
