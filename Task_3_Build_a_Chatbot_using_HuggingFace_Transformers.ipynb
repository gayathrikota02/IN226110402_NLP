## **Step 2**: Importing Libraries

We import the required classes for loading the model and handling text processing.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

## **Step 3**: Loading Pre-trained Model

We load the GPT-2 model from Hugging Face.

### **Why GPT-2?**
- General-purpose language model  
- Generates structured text  
- Works well with prompt-based control  



```python
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token

print("GPT-2 Model loaded successfully!\n")
```

## **Understanding GPT-2**

GPT-2 is a transformer-based language model that generates text by predicting the next word in a sequence.

### **Key Components**:
- Tokenizer → Converts text into numerical tokens  
- Model → Predicts next sequence of words  



## **Step 4:** Chatbot Implementation

In this step, we implement the chatbot logic.

### **Features**:
- Accepts user input  
- Generates responses using GPT-2  
- Maintains conversation history  
- Uses prompt engineering  
- Supports continuous interaction  
- Includes exit condition  

## **Pipeline Flow**

**User Input**

→ Prompt Creation  
→ Tokenization  
→ Model Processing  
→ Response Generation  
→ Decoding  
→ Display Output  
→ Loop until exit  

```python
import datetime

conversation_history = ""

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

while True:
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Conversation ended. Goodbye!")
        break

    # RULE-BASED HANDLING
    # Time query
    if "time" in user_input.lower():
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        print(f"Chatbot: The current time is {current_time}")
        continue

    # Date query
    if "date" in user_input.lower():
        current_date = datetime.datetime.now().strftime("%Y-%m-%d")
        print(f"Chatbot: Today's date is {current_date}")
        continue

    # Greeting
    if user_input.lower() in ["hi", "hello", "hey"]:
        print("Chatbot: Hello! How can I assist you today?")
        continue

    # Conversation memory
    conversation_history += f"User: {user_input}\n"

    prompt = (
        "You are a helpful AI assistant.\n"
        "Answer ONLY what is asked.\n"
        "Be short, clear, and relevant.\n"
        "Do NOT generate extra questions or unrelated content.\n"
        "If unsure, say you are not certain.\n\n"
        f"{conversation_history}"
        "Assistant:"
    )

    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Output generation
    output = model.generate(
        input_ids,
        max_length=input_ids.shape[1] + 60,
        do_sample=False,
        repetition_penalty=1.2,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(
        output[:, input_ids.shape[-1]:][0],
        skip_special_tokens=True
    ).strip()

    response = response.split("User:")[0].strip()

    if not response:
        response = "I'm not completely sure. Could you rephrase your question?"


    conversation_history += f"Assistant: {response}\n"

    print("Chatbot:", response)
```

## **Code Explanation**

### **conversation_history**
Stores previous conversation to maintain context.

Guides GPT-2 to behave like an assistant.

Converts generated tokens into readable text.

Ensures continuous interaction until exit condition.

## **Why the Model Generates Low-Quality or Irrelevant Responses**

The chatbot currently uses a base language model (such as GPT-2), which is not specifically designed for instruction-following or conversational tasks. As a result, the responses may sometimes appear irrelevant, overly generic, or low in quality.

### **Reasons:**

**1. Lack of Instruction Tuning**  
The model is trained on general text data and is designed to predict the next word in a sequence rather than strictly answer questions. It does not inherently understand instructions like “be concise” or “answer accurately.”

**2. No Real-Time Knowledge**  
The model does not have access to real-time data such as the current time, date, or live information. Therefore, it may generate incorrect or assumed answers for such queries.

**3. Text Generation vs Question Answering**  
The model performs text continuation, not true question answering. It generates responses based on learned patterns rather than actual understanding or reasoning.

**4. Limited Context Handling**  
Although basic conversation history is used, the model has limited ability to deeply understand and maintain long conversational context.

**5. Smaller Model Capability**  
Models like GPT-2 are relatively small compared to modern large language models. This limits their ability to generate high-quality, context-aware, and accurate responses.
